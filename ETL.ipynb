# **TRANSFORMACIONES REQUERIDAS:**

##### 1.- Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
amazon_df['id'] = 'a' + amazon_df['show_id']
disney_df['id'] = 'd' + disney_df['show_id']
hulu_df['id'] = 'h' + hulu_df['show_id']
netflix_df['id'] = 'n' + netflix_df['show_id']
```

```python
amazon_df.head()
```

##### 2.- Los valores nulos del campo rating deberán reemplazarse por el string “G” corresponde al maturity rating: “general for all audiences”

```python
amazon_df['rating'].fillna(value ="G", inplace = True)
disney_df['rating'].fillna(value ="G", inplace = True)
hulu_df['rating'].fillna(value ="G", inplace = True)
netflix_df['rating'].fillna(value ="G", inplace = True)
```

```python
amazon_df.head()
```

##### 3.- De haber fechas, deberán tener el formato AAAA-mm-dd

```python
#Utilizamos datetime
amazon_df["date_added"] = pd.to_datetime(amazon_df.date_added)
disney_df["date_added"] = pd.to_datetime(disney_df.date_added)
hulu_df["date_added"] = pd.to_datetime(hulu_df.date_added)
netflix_df["date_added"] = pd.to_datetime(netflix_df.date_added)
```

```python
amazon_df.head()
```

##### 4.- Los campos de texto deberán estar en minúsculas, sin excepciones

```python
#Utilizamos lambda
amazon_df = amazon_df.applymap(lambda s: s.lower() if type(s) == str else s)
disney_df = disney_df.applymap(lambda s: s.lower() if type(s) == str else s)
hulu_df = hulu_df.applymap(lambda s: s.lower() if type(s) == str else s)
netflix_df = netflix_df.applymap(lambda s: s.lower() if type(s) == str else s)
```

```python
amazon_df.head()
```

##### 5.- El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
# Vamos a proceder a separar la columna en dos, para nos aseguramos en limpiar los ESPACIOS EN BLANCO INICIALES
amazon_df['duration'] = amazon_df['duration'].str.lstrip()


disney_df['duration'] = disney_df['duration'].str.lstrip()


hulu_df['duration'] = hulu_df['duration'].str.lstrip()


netflix_df['duration'] = netflix_df['duration'].str.lstrip()
```

```python
# Procedemos a separa la columna en dos partes, a partir del primer espacio.
amazon_df["duration_int"] = amazon_df["duration"].str.split(" ").str.get(0)
amazon_df["duration_type"] = amazon_df["duration"].str.split(" ").str.get(1)

disney_df["duration_int"] = disney_df["duration"].str.split(" ").str.get(0)
disney_df["duration_type"] = disney_df["duration"].str.split(" ").str.get(1)

hulu_df["duration_int"] = hulu_df["duration"].str.split(" ").str.get(0)
hulu_df["duration_type"] = hulu_df["duration"].str.split(" ").str.get(1)

netflix_df["duration_int"] = netflix_df["duration"].str.split(" ").str.get(0)
netflix_df["duration_type"] = netflix_df["duration"].str.split(" ").str.get(1)
```

```python
#comprobamos
amazon_df.info()
```

```python
# CAMBIAMOS el tipo de columna que nos piden para ambos; "duration_int = integer" y "duration_type = string"
# Llenar los valores faltantes con ceros
amazon_df["duration_int"].fillna(0, inplace=True)
disney_df["duration_int"].fillna(0, inplace=True)
hulu_df["duration_int"].fillna(0, inplace=True)
netflix_df["duration_int"].fillna(0, inplace=True)
# Convertir la columna "duration_type" a tipo cadena
amazon_df["duration_int"] = amazon_df["duration_int"].astype(int)
disney_df["duration_int"] = disney_df["duration_int"].astype(int)
hulu_df["duration_int"] = hulu_df["duration_int"].astype(int)
netflix_df["duration_int"] = netflix_df["duration_int"].astype(int)

# Convertir la columna "duration_int" a tipo entero (opcional porque es un object)
amazon_df["duration_type"] = amazon_df["duration_type"].astype(str)
disney_df["duration_type"] = disney_df["duration_type"].astype(str)
hulu_df["duration_type"] = hulu_df["duration_type"].astype(str)
netflix_df["duration_type"] = netflix_df["duration_type"].astype(str)
```

# **UNIFICAMOS LOS ARCHIVOS: amazon, disney, hulu, netflix**

```python
#Unificar 4 plataformas a través de la función “concat” en un dataframe único "general_df", facilitando el desarrollo de las consultas
plataformas_df = pd.concat([amazon_df, disney_df, hulu_df, netflix_df])
```

## RATINGS

```python
# Procedemos a realizar la importación de los archivos csv, de la carpeta ratings 
rating1 = pd.read_csv('ratings/1.csv')
rating2 = pd.read_csv('ratings/2.csv')
rating3 = pd.read_csv('ratings/3.csv')
rating4 = pd.read_csv('ratings/4.csv')
rating5 = pd.read_csv('ratings/5.csv')
rating6 = pd.read_csv('ratings/6.csv')
rating7 = pd.read_csv('ratings/7.csv')
rating8 = pd.read_csv('ratings/8.csv')


#Unifico los 7 archivos csv, ya que contienen el mismo nombre de 

score_df = pd.concat([rating1, rating2, rating3, rating4, rating5, rating6, rating7,rating8], ignore_index=True) 
score_df.info()
```

```python
score_df.head(10)
```

```python
#Cambiamos el formato de la columna "timestamp"
score_df.timestamp= pd.to_datetime(score_df.timestamp, unit="s").dt.date
```

```python
score_df.head(10)
```

Renombro la columna "rating" por "score" para no confundirnos más adelnate

```python
score_df.rename(columns={'rating': 'score'}, inplace=True)
score_df.head()
```

Renombro la columna "moveid" por "id" para poder juntar a la tabla de plataforma con una función en python

```python
score_df.rename(columns={'movieId': 'id'}, inplace=True)
score_df.head()
```

```python
#Todos los datos a minuscula (tal como se hizo en el dataframe de "Plataformas")
score_df = score_df.applymap(lambda x: x.lower() if type(x) == str else x)
score_df.head()
```

```python
#Verificamos si contamos con elementos duplicados en el DataFrame
print( 'El dataframe presenta los siguientes valores duplicados:' , score_df.duplicated().sum())
```

## REALIZAMOS UN PROMEDIO DE LA COLUMNA 'SCORE' Y LA AGRUPAMOS POR PROMEDIO

```python
promedio_score = score_df.groupby('id')['score'].mean()
```

```python
#Redondeamos a dos decimales
promedio_final = promedio_score.round(2)
```

```python
#Realizamos un dataframe pequeño con los promedios con los respectivos id
score_mean = pd.DataFrame({'id': promedio_final.index, 'score_mean': promedio_final.values})
```

```python
score_mean.head()
```

# UNIMOS LOS DATAFRAMES (PLATAFORMAS Y SCORE_MEAN) PARA USO DE LA API

```python
#Unimos medianto el campo 'id'
bd_plataformas= pd.merge(plataformas_df, score_mean, on='id')
```
